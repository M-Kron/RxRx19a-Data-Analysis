Loading the metadata

```python
mdf = pd.read_csv("metadata.csv")
```

combining them into once dataframe

```python
df = df.set_index("site_id")
mdf = mdf.set_index("site_id")
annotated_embeddings = pd.concat([mdf, df], axis=1)
```

```python
annotated_embeddings.head(20)
```

separating sites by experiment

```python
vero_1 = annotated_embeddings[annotated_embeddings.experiment == "VERO-1"]
vero_2 = annotated_embeddings[annotated_embeddings.experiment == "VERO-2"]
hrce_1 = annotated_embeddings[annotated_embeddings.experiment == "HRCE-1"]
hrce_2 = annotated_embeddings[annotated_embeddings.experiment == "HRCE-2"]
```

extracting the untreated sites from each experiment to train the classifier (untreated sites have a treatment value of NaN)

```python
vero_1_training = vero_1[vero_1.treatment != vero_1.treatment]
vero_2_training = vero_2[vero_2.treatment != vero_2.treatment]
hrce_1_training = hrce_1[hrce_1.treatment != hrce_1.treatment]
hrce_2_training = hrce_2[hrce_2.treatment != hrce_2.treatment]
```

creating a series of classifications (1 for viral cells, to be used as positive controls, 0 for mock and irradiated sites, to be used as negative controls) for each training set

```python
vero_1_classes = vero_1_training.disease_condition.map(lambda a: 1 if a == "Active SARS-CoV-2" else 0)
vero_2_classes = vero_2_training.disease_condition.map(lambda a: 1 if a == "Active SARS-CoV-2" else 0)
hrce_1_classes = hrce_1_training.disease_condition.map(lambda a: 1 if a == "Active SARS-CoV-2" else 0)
hrce_2_classes = hrce_2_training.disease_condition.map(lambda a: 1 if a == "Active SARS-CoV-2" else 0)
```

creating and training the classifiers

```python
import sklearn
from sklearn.linear_model import LogisticRegression
```

```python
vero_1_classifier = sklearn.linear_model.LogisticRegression()
vero_2_classifier = sklearn.linear_model.LogisticRegression()
hrce_1_classifier = sklearn.linear_model.LogisticRegression()
hrce_2_classifier = sklearn.linear_model.LogisticRegression()
hrce_1_classifier.max_iter = 1000 #the hrce datasets are considerably larger
hrce_2_classifier.max_iter = 1000
```

```python
vero_1_classifier.fit(vero_1_training.iloc[:, 10:], vero_1_classes) #the first several rows are metadata, hence the slice
vero_2_classifier.fit(vero_2_training.iloc[:, 10:], vero_2_classes)
hrce_1_classifier.fit(hrce_1_training.iloc[:, 10:], hrce_1_classes)
hrce_2_classifier.fit(hrce_2_training.iloc[:, 10:], hrce_2_classes)
```

Annotating each site with it's classification under each of the classifiers (we will mostly work with the the classification from the classifier apropriate to each site's experiment, but having all of the cross-experiment classifications is usefull as well)

```python
classified_metadata = mdf
```

```python
classified_metadata["vero_1_classification"] = df.apply(lambda row: vero_1_classifier.predict_log_proba([row.values])[0][0], axis = 1)
classified_metadata["vero_2_classification"] = df.apply(lambda row: vero_2_classifier.predict_log_proba([row.values])[0][0], axis = 1)
classified_metadata["hrce_1_classification"] = df.apply(lambda row: hrce_1_classifier.predict_log_proba([row.values])[0][0], axis = 1)
classified_metadata["hrce_2_classification"] = df.apply(lambda row: hrce_2_classifier.predict_log_proba([row.values])[0][0], axis = 1)
```
